```python
x = np.linspace(0, 10, 50)
y = np.cos(x) * x + np.random.normal(0, 2, 50)
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
df = pd.DataFrame({"time": x, "observatipon": y})
df.to_csv("time_values.csv", index=False)
```
