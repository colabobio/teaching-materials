# Using packages in Python

A Python package is a [module](https://docs.python.org/3/tutorial/modules.html) of Python code that performs a specific task (such as loading data or creating plots) and it's organized in a way that makes it easy to be shared among other Python users so they can incorporate it in their own projects.

Typically packages implement features that are needed by many people but that are not included in the core of Python - by doing this, developers do not need to be constantly “reinventing the wheel”

There are many packages for Python. The [website of PyPI](https://pypi.org/) (Python Repository Index) is the official third-party software repository for Python, analogous to the CRAN repository for R, where more than half million packages are available. [This tutorial](https://packaging.python.org/en/latest/tutorials/packaging-projects/) explains how to create a Python package and upload it to PyPI for distribution.

## The Pandas and Matplotlib packages

[Pandas](https://pandas.pydata.org/) and [Matplotlib](https://matplotlib.org/) are very popular packages for data handling and visualization.

Pandas provides data structures designed work with "relational" or "labeled" data more easily, while Matplotlib is a comprehensive toolkit for creating static, animated, and interactive visualizations in Python.

### Loading data with pandas

One the pandas package is installed, we can use it to load data from CSV files into our notebooks, and then work with the data in different ways. First of all, we need to import the pandas package into our notebook, so we are able to use it. In the cell below, we import the pandas package and also give the "pd" alias to it so it's easier to write the code afterwards.

````python
import pandas as pd
````

The function [read_csv()](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) in pandas takes the name of the CSV file to load and it outputs a "data frame" object that contains all the rows and columns in the file, by 

````python
housing_data = pd.read_csv("Housing.csv")
housing_data # this will be considered the output of the cell, so it will just print the contents of the data frame
````

A pandas data frame contains columns, which are the variables in the data, and rows, which are each one of the entries (or observations) in the data. We can get a summary description of the data frame by using the [describe()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html) function:

````python
housing_data.describe(include='all')
````

Other utility functions allow us to get the number of elements in the data frame:

````python
print(housing_data.shape) # shape is a tuple with the number of rows and columns
print(len(housing_data))  # len() calculated on the entire data frame returns the number of rows
print(housing_data.size)  # size is the the number of rows multiplied by the number of columns
````

The data frame object contains a list with the name of all the columns, precisely called columns:

````python
print(housing_data.columns)
print(len(housing_data.columns))
````

### Accessing individual columns in pandas

To access a specific column in the data frame, we only need to put its name between square brackets, this will return all the values in the column:

````python
housing_data['area']
````

We can apply several functions on a column, like so:

````python
print('Description of the area variable')
print(housing_data['area'].describe())
print('    ')
print(f'Mean area = {housing_data['area'].mean():.2f}')
print(f'Standard deviation of area = {housing_data['area'].std():.2f}')
````

### Getting a subset of the data that matches a condition

A very common operation with data is to select a subset of a table based on a condition applied to a sepecific column or variable. For instance, if we want to select all the rows in our housing data that correspond to houses with an area lower than 3000 square feet, we can do the following:

````python
housing_data[housing_data['area'] < 3000]
````

We can also assing the result of this selection to a new data frame object and to calculations only on that data frame:

````python
small_houses = housing_data[housing_data['area'] < 3000]
print(f'Mean price of houses under 3,000 square feet is {small_houses['price'].mean():,.2f} USD') # We can use "," to format the number!
````

We can apply several conditions at once, for example to select houses under 3,000 square feet and 5,000,000 USD we can write the following:

````python
# Notice that each condition is enclosed in parenthesis and 
# the 'AND' logic operation is represented by the '&' character
small_cheaper_houses = housing_data[(housing_data['area'] < 3000) & (housing_data['price'] < 5000000)]
small_cheaper_houses
````

Finally, we can also use slicing in pandas data frames to select a range of rows:

````python
some_houses = housing_data[0:20]
some_houses
````

The individual values for each element in the table can be accessed from the values array:

````python
some_houses.values
````

````python
# This is the price of the first house, since is the 0th row and the 0th column
print(some_houses.values[0][0])

# This is the area of the second house, since is the 1th row and the 1th column
print(some_houses.values[1][1])
````

### Plotting data with Matplotlib

Similar to pandas, once Matplotlib is installed, then we can import it (specifically its sub-module for plotting in Python) with the following line:

````python
import matplotlib.pyplot as plt
````

In this case, we are using the 'plt' alias to avoid having to matplotlib.pyplot every time we need to use its functions. For example, we can start by plotting the histogram of the house prices, with the [hist()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html) function:

````python
plt.figure(figsize=(5,5))                  # Set the figure size (optional) to 5x5 inches
plt.hist(housing_data['price'], bins=20)   # Give hist() the data to plot, and the number of bins to use
plt.xlabel('Price')                        # Set the label for the x axis (optional)
plt.ylabel('Count')                        # Set the label for the y axis (optional)
plt.tight_layout()                         # Set a tight layout for properly saving the figure (optional)
plt.savefig("price_hist.pdf")              # Save plot image to pdf file (optional)
````

Most functions are self-explanatory, but we can see all their parameters in the online docs: [figure()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html), [xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html), [ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html). To save a plot as a figure in a file we have the [savefig()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html) function, which automatically writes the file in the correct format according to the file extension we provide. The [tight_layout()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.tight_layout.html) function is an easy way to adjust the spacing around the axes in a plot so all the labels are correctly displayed in the image file. 

In a very similar way, we can generate histograms of other variables in the data, scatter plots between two variables in the data, box plots and violin plots of individual variables to get visualize the main features of their distributions:

````python
plt.figure(figsize=(5,5))
plt.hist(housing_data['area'], bins=20)
plt.xlabel('area')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig("area_hist.pdf")
````

````python
plt.figure(figsize=(5,5))
# For the scatter plot, we need to provide two variables, the first for X and the second for Y:
plt.scatter(housing_data['price'], housing_data['area'])  
plt.xlabel('Price')
plt.ylabel('Area')
plt.tight_layout()
plt.savefig("price_area_scatter.pdf")
````

````python
plt.figure(figsize=(5,5))
plt.boxplot(housing_data['area'])
plt.xlabel('Area')
plt.tight_layout()
plt.savefig("area_boxplot.pdf")
````

````python
plt.figure(figsize=(5,5))
plt.violinplot(housing_data['area'])
plt.xlabel('Area')
plt.tight_layout()
plt.savefig("area_violinplot.pdf")
````

### Making line plots with Matplotlib

The housing data set does not lend itself to demonstrate line plots, so we can use synthetic data of values changing over time:

````python
time_data = pd.read_csv('time_values.csv')
time_data
````

A basic line plot does not need any parameters or settings, in which case it Matplotlib will use its defauts:

````python
plt.plot(time_data['time'], time_data['value'])
````

One simple setting for the line plot is the color of the line. We can use ["named colors"](https://matplotlib.org/3.1.0/gallery/color/named_colors.html) in Matplotlib that are easy to remember

![Matplotlib named colors](https://matplotlib.org/stable/_images/sphx_glr_named_colors_003_2_00x.png)

````python
plt.plot(time_data['time'], time_data['value'], color='orange')
````

Another alternative are the ["XKCD colors"](https://xkcd.com/color/rgb/), which resulted from a [public survey](https://blog.xkcd.com/2010/05/03/color-survey-results/), and are not coincident with the default named colors in Matplotlib:

````python
plt.plot(time_data['time'], time_data['value'], color='xkcd:orange')
````

The colors in a Matplotlib can also be set using [Hexadecimal (HEX) color values](https://www.w3schools.com/colors/colors_hexadecimal.asp), which are also used in HTML:

````python
plt.plot(time_data['time'], time_data['value'], color='#8e82fe')
````

### Dealing with datetime values

Data can also contain values to represent dates or times of the day (i.e.: "2023-10-01" or "10:04:03") but pandas will assume that they are just strings, so we would need to conver them to the correct type. For example, the following data contains a column where the date values are in the format "yyy-mm-dd" so we need to convert them into datetime values after loading the file into our notebook:

````python
df = pd.read_csv('DailyDelhiClimateData.csv')
df
````

````python
df.describe(include='all')
````

````python
df.dtypes
````

The conversion can be done with the [to_datetime()](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) function. We can specify the format in which the datetime data is stored, but the default is "yyy-mm-dd" so because of that we don't need to set any arguments in this case: 

````python
df['date'] = pd.to_datetime(df['date'])
````

We can now plot the data using date as the x values for a line plot:

````python
plt.plot(df['date'], df['meantemp'])
plt.xticks(rotation=90)
````

We can adjust the settings in the xticks() function so the tick labels are rotated by an angle and right-adjusted:

````python
plt.figure(figsize=(10,5))
plt.plot(df['date'], df['meantemp'], color='darkorchid')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Date')
plt.ylabel('Mean temperature')
plt.tight_layout()
plt.savefig("delhi_meantemp_plot.pdf")
````
