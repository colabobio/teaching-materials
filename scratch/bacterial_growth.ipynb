````python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
````

## Formula for bacterial growth including survival rate

If n0 is he initial population, and s is the survival rate, then:

```
n1 = s * n0 * 2

n2 = s * n1 * 2 = s * (s * n0 * 2) * 2 = n0 * (s * 2)^2

...

nt = n0 * (s * 2) ^ t
```

````python
reproduction_rate = 2

surv_rate_mean = 0.7
surv_rate_std = 0.1

inital_pop = 10
````

````python
x = np.linspace(surv_rate_mean - 3 * surv_rate_std, surv_rate_mean + 3 * surv_rate_std, 100)
fig = plt.figure(figsize = (4, 3))
plt.plot(x, stats.norm.pdf(x, surv_rate_mean, surv_rate_std))
plt.show()
````

````python
fig = plt.figure(figsize = (10, 6))

t = np.linspace(0, 20, 100)

surv_rates = np.random.normal(surv_rate_mean, surv_rate_std, 5)
surv_rates.sort()
for s in surv_rates[::-1]:
    n = inital_pop * (s * reproduction_rate) ** t
    plt.plot(t, n, label=f'Survival rate {int(100 * s)}%')
plt.xlabel("time")
plt.ylabel("Population size")
plt.legend()
plt.show()
````
