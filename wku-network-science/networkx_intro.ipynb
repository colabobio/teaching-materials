```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
print(G)
```

```python
print(G.nodes())
print(G.edges())
```

```python
G.add_node("A")
```

```python
print(G)
```

```python
print(G.nodes())
```

```python
G.add_nodes_from(["B", "C", "D", "E"])
```

```python
print(G.nodes())
```

```python
print(G.edges())
```

```python
G.add_edge(*("A", "B"))
```

```python
print(G.edges())
```

```python
G.add_edges_from([("A", "C"), ("B", "D"), ("B", "E"), ("C", "E")])
```

```python
print(G.edges())
```

```python
print(G)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G)
plt.show() 
```

```python
nx.draw(G)
plt.savefig("graph.png") 
```

```python
nx.draw(G, with_labels=True, font_weight="bold", node_color='r', edge_color='b')
plt.show()
```
