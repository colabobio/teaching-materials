```python
import networkx as nx
import matplotlib.pyplot as plt
import pickle
```

```python
with open('full-network.pickle', 'rb') as f:
    Gfull = pickle.load(f)
```

```python
print(Gfull)
```

```python
Gfull.degree()
```

```python
nx.draw(Gfull, node_size=10, width=0.5)
plt.show()
```

```python
pos = nx.spring_layout(Gfull)
plt.figure(figsize=(15, 15))
nx.draw_networkx(Gfull, pos, with_labels=False, node_size=10, width=0.1, alpha=0.5)
plt.show()
```
